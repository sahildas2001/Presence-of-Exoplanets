```python
def normalize_negative_one(arr):
    normalized_input = (arr - np.amin(arr)) / (np.amax(arr) - np.amin(arr))
    return 2 * normalized_input - 1

normalized = np.zeros(flux_vals.shape)

for i in range (0, 5087):
    normalized[:][i] = normalize_negative_one(flux_vals[:][i])
```

```python
fig = plt.figure(figsize=(15,15))
for i in range(0, 16):
    plt.subplot(4, 4, i+1)
    plt.plot(normalized[:][i])

plt.show()
```
